```python
import numpy as np
import matplotlib.pyplot as plt
import math
from math import radians
```

```python
def simple_calculator():
    print("you have selected normal calculator ")
    print("you can perform addition, subtraction , multiplication and division")
    first=float(input("enter the value of first number="))
    operator=input("enter the type of operator you want=")
    second=float(input("enter the value of second number="))
    if(operator=='+'):
        result=first+second
        print("the value of your result =",result)
    elif(operator=='-'):
        result=first-second
        print("the value of your result =",result)
    elif(operator=='*'):
        result=first*second
        print("the value of your result =",result)
    elif(operator=='/'):
        result=first/second
        print("the value of your result =",result)
    else:
        print("you have entered some wrong input")
```

```python
def power():
    print("you have selected indices")
    print("you can do squaring of a number cube , square root and cube root")
    a=input("do you want power or root of a number=")
    if(a=='power'):
        first=float(input("enter the value of number ="))
        operator=int(input("enter the power of your number="))
        result=first**operator
        print(result)
    elif(a=='root'):
        first=float(input("enter the value of number ="))
        operator=input("enter which type or root do you want square root or cube root=")
        if(operator=='square root'):
            result=np.sqrt(first)
            print("the value of your result =",result)
        elif(operator=='cube root'):
            result=np.cbrt(first)
            print("the value of your result =",result)
        else:
            print("sorry , we are unable to find your result")
```

```python
def geometry():
    print("you have selected one of the application of maths")
    print("you can find area and volume of certain shapes")
    a=input("do you want to find area and volume =")
    if(a=='area'):
        print("we can find the area of ")
        print("circle, square, rhombus, rectangle, parallelogram, triangle")
        b=input("enter the name of shape=")
        if(b=='circle'):
            first=float(input("enter the value of radius="))
            result=3.14*first*first
            print("the value of your result =",result)
        elif(b=='square' or b=='rhombus'):
            first=float(input("enter the value length of side="))
            result=first*first
            print("the value of your result =",result)
        elif(b=='rectangle' or b=='parallelogram' ):
            first=float(input("enter the value of length="))
            second=float(input("enter the value of breadth="))
            result=first*second
            print("the value of your result =",result)
        elif(b=='triangle'):
            base=float(input("enter the value of base="))
            height=float(input("enter  the value of height="))
            result=base*height/2
            print("the value of your result =",result)
        else:
            print("you have given wrong input")
    if(a=='volume'):
        print("we can find the volume of ")
        print("cube, cuboid, sphere, cone, cylinder, prism, pyramid")
        b=input("enter the name of the shape=")
        if(b=='cube'):
            length=float(input("enter the length of the cube="))
            result=length*length*length
            print("the value of your result =",result)
        elif(b=='cuboid'):
            length=float(input("enter the value of length="))
            breadth=float(input("entrer the value of breadth="))
            height=float(input("enter the value of height="))
            result=length*breadth*height
            print("the value of your result =",result)
        elif(b=='sphere'):
            radius=float(input("enter the value of radius="))
            result=4*3.14*radius*radius*radius/3
            print("the value of your result =",result)
        elif(b=='cone'):
            radius=float(input("enter the value of radius="))
            height=float(input("enter the value of height of a cone="))
            result=3.14*radius*radius*height/3
            print("the value of your result =",result)
        elif(b=='cylinder'):
            radius=float(input("enter the radius of a cylinder="))
            height=float(input("enter the value of height="))
            result=3.14*radius*radius*height
            print("the value of your result =",result)
        elif(b=='prism'):
            base=float(input("enter the value of base"))
            height=float(input("enter the value of height"))
            result=base*height
            print("the value of your result =",result)
        elif(b=='pyramid'):
            base=float(input("enter the value of base="))
            height=float(input("enter the value of height="))
            result=base*height/3
            print("the value of your result =",result)
        else:
            print("wrong input")
```

```python
def angles():
    print("you have selected an application of maths")
    print("we can calculate any one interior angle of polygon")
    print("polygon= triangle, rectangle, pentagon, hexagon")
    a=input("enter the type of polygon you want=")
    if(a=='triangle'):
        first=float(input("enter the value of first angle="))
        second= float(input("enter the value of second angle="))
        result=180-first-second
        print("the value of the third angle =",result)
    elif(a=='rectangle' or a=='square' or a=='parallelogram' or a=='rhombus'):
        first=float(input("enter the value of first angle="))
        second= float(input("enter the value of second angle="))
        third=float(input("enter the value of third angle="))
        result= 360 - first - second- third
        print("the value of fourth angle =",result)
    elif(a=='pentagon'):
        first=float(input("enter the value of first angle="))
        second= float(input("enter the value of second angle="))
        third=float(input("enter the value of third angle="))
        fourth=float(input("enter the value of fourth angle="))
        result= 540 -first-second-third-fourth
        print("the value of fifth angle =",result)
    elif(a=='hexagon'):
        first=float(input("enter the value of first angle="))
        second= float(input("enter the value of second angle="))
        third=float(input("enter the value of third angle="))
        fourth=float(input("enter the value of fourth angle="))
        fifth=float(input("enter the value of fifth angle="))
        result=720 - first - second - third - fourth - fifth
        print("the value of sixth angle =",result)
```

```python
def graph():
    print("you are interested to plot a graph")
    print("you can plot a graph of sin(x) ,cos(x) ,tan(x),cot(x),sec(x),cosec(x)")
    x=np.linspace(0,20,100)
    a=input("enter the function of which you want to plot a graph=")
    if(a=='sin(x)'):
        y=np.sin(x)
        plt.xlabel("angle")
        plt.ylabel("value")
        plt.title("sine wave")
        plt.plot(x,y)
        plt.show()
    elif(a=='cos(x)'):
        y=np.cos(x)
        plt.xlabel("angle")
        plt.ylabel("value")
        plt.title("cosine wave")
        plt.plot(x,y)
        plt.show()
    elif(a=='tan(x)'):
        y=np.tan(x)
        plt.xlabel("angle")
        plt.ylabel("value")
        plt.title("tan curve")
        plt.plot(x,y)
        plt.show()
    elif(a=='cot(x)'):
        y=np.cos(x)/np.sin(x)
        plt.xlabel("angle")
        plt.ylabel("value")
        plt.title("cot curve")
        plt.plot(x,y)
        plt.show()
    elif(a=='sec(x)'):
        y=1/np.cos(x)
        plt.xlabel("angle")
        plt.ylabel("value")
        plt.title("sec curve")
        plt.plot(x,y)
        plt.show()
    elif(a=='cosec(x)'):
        y=1/np.sin(x)
        plt.xlabel("angle")
        plt.ylabel("value")
        plt.title("cosec curve")
        plt.plot(x,y)
        plt.show()
```

```python
def avg():
    print("you have selected to find the average of two number")
    first=float(input("enter the value of first number"))
    second=float(input("enter the value of second number"))
    result=(first+second)/2
    print("the value of your result =",result)
```

```python
def factorial():
    print("you have selected factorial")
    print("you can find factorial of any number")
    factorial=1
    num=int(input("enter the value to find factorial="))
    if(num<0):
        print("factorial doesn’t exist")
    elif(num==0):
        result=1
        print("the value of factorial =",result)
    else:
        for i in range(1,num+1):
            factorial=factorial*i
        print("the value of factorial =",factorial)
```

```python
def trigo():
    print("you have selected trigonometry")
    print("you can find the trigonometric value of any angle")
    a=float(input("enter the value of angle="))
    b=radians(a)
    operator=input("enter the trigonometric  function")
    if(operator=='sin'):
        result=np.sin(b)
        print("the value of your result =",result)
    elif(operator=='cos'):
        result=np.cos(b)
        print("the value of your result =",result)
    elif(operator=='tan'):
        result=np.tan(b)
        print("the value of your result",result)
    elif(operator=='cot'):
        result=1/np.tan(b)
        print("the value of your result",result)
    elif(operator=='sec'):
        result=1/np.cos(b)
        print("the value of your result",result)
    elif(operator=='cosec'):
        result=1/np.sin(b)
        print("the value of your result =",result)
```
